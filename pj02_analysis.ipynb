# Hypothesis:
## I think that there is a strong correlation between the percent of females enrolled in primary school and the percent of females in the labor force.


### I'm going to read in the CSV file into a DataFrame and then I initialized two named constants for each of the indicator code strings I'm going to explore.

```python
In [1]: import pandas as pd
In [2]: df = pd.read_csv("../../data/databank_education_2018.csv")
In [3]: SCHOOL = "SE.PRM.ENRR.FE"
In [4]: LABOR_FORCE = "SL.TLF.TOTL.FE.ZS"
```

### Next, I printed the amount of rows and columns the DataFrame has and then I used the head method to display the first 10 rows.

```python
In [5]: print(df.shape)
In [6]: df.head(10)
```

### Next, I selected the "Country" column and the two indicator columns I chose. I then printed out the first 10 rows.

```python
In [7]: columns = df[["Country", SCHOOL, LABOR_FORCE]]
In [8]: columns.head(10)
```

### Next, I only selected the rows that had values that were numbers. I then printed out the number of rows and columns this data set has. Finally, I printed out the first 10 rows of the data set.

```python
In [9]: data = columns[columns[SCHOOL].notna() & columns[LABOR_FORCE].notna()]
In [10]: print(data.shape)
In [11]: data.head(10)
```

### Next, I sorted the data by the percent of the labor force that is female. I then printed out the first 10 rows.

```python
In [12]: data.sort_values(by=LABOR_FORCE, ascending=False).head(10)
```

### Next, I created a scatter plot with a regression line. I made the percent of females enrolled in primary school be the x-axis and I made the percent of females in the labor force be the y-axis.

```python
In [13]: import matplotlib.pyplot as plt
In [14]: import numpy as np
In [15]: data.plot.scatter(x=SCHOOL, y=LABOR_FORCE, alpha=0.5, xlabel="percent of females enrolled in primary school", ylabel="percent of females in the labor force", title="females enrolled in primary school vs. females in the labor force")
m, b = np.polyfit(data[SCHOOL], data[LABOR_FORCE], 1)
plt.plot(data[SCHOOL], m*data[SCHOOL] + b)
```

### Next, I found the correlation between the percent of females enrolled in primary school and the percent of females in the labor force.

```python
data[SCHOOL].corr(data[LABOR_FORCE], method="pearson")
```

# Conclusion:
I was not able to prove that there is a strong correlation between the percent of females enrolled in primary school and the percent of females in the labor force. The correlation that I found is potitive, however since it's only about 0.141, it isn't very strong.
